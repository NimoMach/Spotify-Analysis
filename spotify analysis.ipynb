# Extracting only top200 valued one's in chart column

## as our main focus is related to streams lets only explore that part of the dataset

```python
top200_streams=df[df['chart']=='top200']
top200_streams.head()
```

```python
top200_streams.shape
```

```python
#understanding if there are null values in top200 df 
#len(most_streams[most_streams['streams'].isna()==True])
top200_streams['streams'].isna().head()
```

```python
top200_streams['region'].unique()
```

# Removing global region records from top200_streams dataframe

```python
top200_streams=top200_streams[top200_streams['region']!='Global']
```

# Getting the results of regional wise streams per year.

```python
region_streams=top200_streams.groupby(['year','region'],as_index=False)['streams'].sum()
region_streams.head()
#this resulted data set is used for visualization-1
```

# lets look at the most popular artist based on streams (most repeatedly streamed artist)

```python
top200_streams['artist'].value_counts().sort_values(ascending=False).head()
```

# getting details related to top artist - ed sheeran

```python
ed_sheeran=top200_streams[top200_streams['artist'].str.contains('Ed Sheeran')]
ed_sheeran.head()
```

## We are even considering the collaborations with other artists

```python
ed_sheeran['artist'].unique()
```

```python
#Regional wise streams for only Ed Sheeran and his collaborations
edsheeran_streams=ed_sheeran.groupby(['year','region'],as_index=False)['streams'].sum()
edsheeran_streams.head()
#this resulted data set is used for visualization-2
```

# getting the top songs of ed sheeran and their trends

```python
top_songs=ed_sheeran.value_counts('title').sort_values(ascending=False)
top_songs
```

```python
#Now lets only focus on the top 3 songs and try finding insights related to them.

top_3_song=ed_sheeran[ed_sheeran['title'].isin(['Photograph','Shape of You','Perfect'])]
top_3_song.head()
```

```python
#The top 3 songs streams per year
top_song_analysis=top_3_song.groupby(['year','title'],as_index=False)['streams'].sum()
top_song_analysis.head()
#this resulted data set is used for visualization-3
```
